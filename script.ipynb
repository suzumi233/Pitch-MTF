```python
import os
import shutil

def find_and_copy_wav_files(root_dir, output_dir):
    for root, dirs, files in os.walk(root_dir):
        for file in files:
            if file.endswith(".wav"):
                source_path = os.path.join(root, file)
                destination_path = os.path.join(output_dir, file)
                shutil.copyfile(source_path, destination_path)
                print("Copied:", source_path)

# 示例用法
dataset_root = '/Users/suzumi/Downloads/timitfull'
output_dir = '/Users/suzumi/Downloads/timitwav'

# 创建输出文件夹
os.makedirs(output_dir, exist_ok=True)

# 复制所有的 .wav 文件到输出文件夹
find_and_copy_wav_files(dataset_root, output_dir)
```

```python
import os

def rename_files(root_dir):
    count = 1
    for root, dirs, files in os.walk(root_dir):
        for file in files:
            file_ext = os.path.splitext(file)[1]  # 获取文件扩展名
            new_name = f"tmt{str(count).zfill(4)}{file_ext}"  # 构建新文件名
            old_path = os.path.join(root, file)  # 原始文件路径
            new_path = os.path.join(root, new_name)  # 新文件路径
            os.rename(old_path, new_path)  # 重命名文件
            print("Renamed:", old_path, "to", new_path)
            count += 1

# 示例用法
dataset_root = '/Users/suzumi/Downloads/timitwav'

# 重命名数据集中的文件
rename_files(dataset_root)
```

```python
import os
import wave

def check_sample_rate(root_dir):
    for root, dirs, files in os.walk(root_dir):
        for file in files:
            if file.endswith(".wav"):
                file_path = os.path.join(root, file)
                with wave.open(file_path, 'r') as wav_file:
                    sample_rate = wav_file.getframerate()
                    if sample_rate != 16000:
                        print("Invalid sample rate:", file_path, "Sample rate:", sample_rate)
                    else:
                        print("Valid sample rate:", file_path)

# 示例用法
dataset_root = '/Users/suzumi/Downloads/timitwav'

# 检查数据集中的采样率
check_sample_rate(dataset_root)
```

```python
import os
import wave
import soundfile as sf

def resample_files(root_dir, target_sample_rate):
    for root, dirs, files in os.walk(root_dir):
        for file in files:
            if file.endswith(".wav"):
                file_path = os.path.join(root, file)
                output_dir = os.path.join(root, "resampled")
                os.makedirs(output_dir, exist_ok=True)  # 创建输出目录
                output_path = os.path.join(output_dir, file)
                data, current_sample_rate = sf.read(file_path)
                if current_sample_rate != target_sample_rate:
                    resampled_data = sf.resample(data, target_sample_rate)
                    sf.write(output_path, resampled_data, target_sample_rate)
                    print("Resampled:", file_path, "Saved as:", output_path)
                else:
                    print("No resampling needed:", file_path)

# 示例用法
dataset_root = '/Users/suzumi/Downloads/timitwav'
target_sample_rate = 16000

# 将数据集中的音频文件采样率统一为 16000
resample_files(dataset_root, target_sample_rate)
```
